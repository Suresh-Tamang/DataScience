## CREDIT CARD FRAUD DETECTION

```python
# importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# loading dataset
df = pd.read_csv('creditcard.csv')
df.head()
```

Statistical vaues 

```python
df.describe()
```

Checking the null values 

```python
df.isnull().sum()
```

we can see the dataset is very clean 

Extracting information about all columns


```python
df.info()
```

## Data visualization

```python
fit = plt.figure(figsize=(8,6))
sns.countplot(x = 'Class', data = df)
plt.xlabel('Fraud class 1:Fraud, 0:Not Fraud')
plt.ylabel('Count')
plt.title("Fraud Classification Count")
plt.show()
```

## Setting Up Dataset For Model Development

```python
X = df.drop(['Class'],axis = 1)
X.head()
```

```python
Y = df['Class']
Y.head()
```

### Splitting data in the train and test set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state = 5)
```

## Model Development

### KNeighbor-Nearest Classification 


```python
from sklearn.neighbors import KNeighborsClassifier
neighbor = KNeighborsClassifier(n_neighbors = 5)
```

Training tahe model

```python
neighbor.fit(x_train,y_train)
```

Making predictions

```python
prediction1 = neighbor.predict(x_test)
```

```python
print(prediction1)
```

Model Evaluation

```python
from sklearn.metrics import confusion_matrix,accuracy_score
accuracy = accuracy_score(y_test,prediction1)
conMatrix = confusion_matrix(y_test,prediction1)
```

```python
print("Accuracy score is {:.2f}%".format(accuracy*100))
```

```python
print("ConfusionMatrix:\n",conMatrix)
```

The Model is 99.82% accurate to detect the fraud.

## LogisticRegression Classification Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression(C = 0.001,solver = 'liblinear')
```

Fitting the data


```python
LR.fit(x_train,y_train)
```

Making prediction 

```python
LRpredict = LR.predict(x_test)
print(LRpredict)
```

Model Evaluation

```python
print("Accuracy of the Model is  {:.2f}%.".format(accuracy_score(y_test,LRpredict)*100))
```

```python
print("ConfusionMatrix :\n",confusion_matrix(y_test,LRpredict))
```

This model gives more accurate prediction than KNN model

## DecisionTree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
import sklearn.tree as tree
```

```python
DST = DecisionTreeClassifier()
```

fitting the model


```python
DST.fit(x_train,y_train)
```

Prediction using test data set

```python
DSprediction = DST.predict(x_test)
```

```python
print(DSprediction)
```

Model Evaluation

```python
print("Confusion Matrix:")
confusion_matrix(y_test,DSprediction)
```

```python
print('Accuracy of the model is {:.2f}%.'.format(accuracy_score(y_test,DSprediction)*100))
```

# Conclusions

From the above models the most accurate model is decision tree model so DecisionTreeClassifier is the best predictor for Fraud Detection in Credit card.
